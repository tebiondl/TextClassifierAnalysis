```python
import os
import matplotlib.pyplot as plt
```

```python
base_dir = 'new_categories'
file_counts = []

for i in range(5):
    subfolder = os.path.join(base_dir, f'category_{i}')
    file_count = len([name for name in os.listdir(subfolder) if os.path.isfile(os.path.join(subfolder, name))])
    file_counts.append(file_count)
```

```python
plt.bar(range(5), file_counts)
plt.xlabel('Subcarpeta')
plt.ylabel('Número de archivos')
plt.title('Recuento de archivos por subcarpeta')
plt.show()
```
