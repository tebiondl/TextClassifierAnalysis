```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#xlnet-base-cased, bert-base-uncased
df_xlnet = pd.read_csv("results_xlnet-base-cased.txt", delimiter=":", names=["Metric", "Value"])
df_bert = pd.read_csv("results_bert-base-uncased.txt", delimiter=":", names=["Metric", "Value"])
```

```python
metric = "F1"
```

```python
accuracy_xlnet = df_xlnet[df_xlnet['Metric'].str.strip() == metric]['Value'].astype(float)
accuracy_bert = df_bert[df_bert['Metric'].str.strip() == metric]['Value'].astype(float)
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([accuracy_xlnet, accuracy_bert], notch=True, patch_artist=True, showmeans=True)
plt.xticks([1, 2], ['XLNet', 'BERT'])
plt.title("Comparaci√≥n de "+metric+" entre XLNet y BERT")
plt.ylabel(metric)
plt.grid(True)
plt.show()
```
